### Example: Wireless connection

<div class="boxed">
Radio waves are a form of eletctromagnetic radiation with wave forms between 30 Hz and 300 GHz.

- Generated by oscillating electic fields
- In the receiver the radio singla create a tiny voltage which can be measured




```python
f = 1 # frequency
A = 1 # amplitude
phi = 0 # phase
y = lambda t: A*np.sin(2*np.pi*f*t+phi)

# Data for plotting
t = np.arange(0.0, 2.0, 0.01)

fig, ax = plt.subplots()
ax.plot(t, y(t))

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Sine wave')
ax.grid()
plt.show()
```

Here:
- A: amplitude
- f: frequency, the number of oscillations (in Hz)
- ω = 2πf: angular frequency
- φ: phase, specifies position at t = 0

### Hidden Terminal problem 
- Hidden terminals are senders that cannot sense each other but nonetheless collide at intended receiver

<img src="hidden.png"></img>

- B, A hear each other
- B, C hear each other
- A, C can not hear each other means A, C unaware of their interference at B

```python
z = lambda t: A*np.sin(2*np.pi*f*t+1)

fig, ax = plt.subplots()
ax.plot(t, y(t))
ax.plot(t, z(t))
ax.plot(t, y(t)+z(t))
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Two interfering signals')
ax.grid()
plt.show()
```

Protocol solutions exist to the hidden terminal problem
- e.g. Request To Send/Clear To Send (RTS/CTS)

### Signal Attenuation

<img src="attenuation.png"></img>

- B, A hear each other
- B, C hear each other
- A, C can not hear each other interfering at B



```python
# signal with attenuation
A = lambda t: 1/(t+1)/(t+1)
f = 10
z = lambda t: A(t)*np.sin(2*np.pi*f*t+1)

fig, ax = plt.subplots()
ax.plot(t, z(t))
ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='Attenuation')
ax.grid()
plt.show()
```
